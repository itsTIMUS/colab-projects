```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the data set into pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.shape
```

```python
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
```

```python
#distribution of legal and fraud transaction
credit_card_data['Class'].value_counts() #this data set is highly unbalanced
```

```python
#seperating the data for ananlysis
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape,fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values of both transaction
credit_card_data.groupby('Class').mean()
```

```python
#build a sample dataset containing a similar distribution of fraud and normal transac
legit_sample = legit.sample(n=211)
```

```python
legit_sample.shape
```

```python
#concatenate 2 data frames
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset['Class'].value_counts()
```

```python
#compare the values of both transaction
new_dataset.groupby('Class').mean()
```

```python
#seperating data and labels
X = new_dataset.drop(columns=['Class'], axis=1)
Y = new_dataset['Class']
print(X,Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

```python
#model traing using logistic regression
model = LogisticRegression()
#training the model with training the data
model.fit(X_train,Y_train)
```

```python
#train model evaluation
X_train_prediction = model.predict(X_train)
Training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print('train accuracy score',Training_data_accuracy)
```

```python
# test model evaluation
X_test_prediction = model.predict(X_test)
Test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print('test accuracy score',Test_data_accuracy)
```

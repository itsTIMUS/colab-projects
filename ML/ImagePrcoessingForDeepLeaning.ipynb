```python
#Resizing the images with pillow library
from PIL import Image
```

```python
img = Image.open('/content/dog1.jpeg')
```

```python
img_resized = img.resize((200,200))
```

```python
img_resized.save('dog_img_resized.jpeg')
```

```python
#loading a image through matplotlib.image module
img_res = mpimg.imread('/content/dog_img_resized.jpeg')
img_plot = plt.imshow(img_res)
plt.show()
```

```python
# rgb to grayscale using OpenCV
import cv2
```

```python
img = cv2.imread('/content/dog1.jpeg')
type(img)
```

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
type(grayscale_img)
```

```python
print(grayscale_img.shape)
```

```python
#cv2.imshow will display the image but it will not work in colab instead
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(grayscale_img)
```

```python
#saving the image
cv2.imwrite('dog_grayscale_img.jpeg', grayscale_img)
```

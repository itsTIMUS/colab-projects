```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie_data = pd.read_csv('/content/movies.csv')
```

```python
movie_data.head()
```

```python
movie_data.shape
```

```python
selected_feature = ['genres','keywords','tagline','cast','director']
print(selected_feature)
```

```python
#replacing null values with null string
for feature in selected_feature:
  movie_data[feature] = movie_data[feature].fillna('')
```

```python
#combining all 5 features
combined_feature = movie_data['genres']+' '+movie_data['keywords']+' '+movie_data['tagline']+' '+movie_data['cast']+' '+movie_data['director']
print(combined_feature)
```

```python
#converting the text data into feature vectors
vectorizer = TfidfVectorizer()
feature_vector = vectorizer.fit_transform(combined_feature)
print(feature_vector)
```

```python
#cosine similarity
# getting the similaruty score
similarity = cosine_similarity(feature_vector)
print(similarity)
```

```python
similarity.shape
```

```python
#getting the movie name from the user
movie_name = input('Enter your favourite movie name: ')
```

```python
#creaating a list with all the movie names in the dataset
list_of_all_title = movie_data['title'].tolist()
print(list_of_all_title)
```

```python
#finding the close match to the movie name given by the user
finding_match =  difflib.get_close_matches(movie_name,list_of_all_title)
print(finding_match)
```

```python
close_match = finding_match[0]
print(close_match)
```

```python
from collections.abc import ValuesView
#find the index of the movie with title
index_of_movie = movie_data[movie_data.title == close_match]['index'].values[0]
print(index_of_movie)
```

```python
# getting a list of similar values
similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the values based on similarity score
sorted_similar_movies = sorted(similarity_score,key = lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#printing the name of the movies
print('movies suggested for you: \n')
i=1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movie_data[movie_data.index==index]['title'].values[0]
  if(i<30):
    print(i,'.',title_from_index)
    i+=1
```

```python
#movie recommendation system
movie_name = input('Enter your favourite movie name: ')
list_of_all_title = movie_data['title'].tolist()
finding_match =  difflib.get_close_matches(movie_name,list_of_all_title)
close_match = finding_match[0]
index_of_movie = movie_data[movie_data.title == close_match]['index'].values[0]
similarity_score = list(enumerate(similarity[index_of_movie]))
sorted_similar_movies = sorted(similarity_score,key = lambda x:x[1],reverse=True)
#printing the name of the movies
print('movies suggested for you: \n')
i=1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movie_data[movie_data.index==index]['title'].values[0]
  if(i<30):
    print(i,'.',title_from_index)
    i+=1
```
